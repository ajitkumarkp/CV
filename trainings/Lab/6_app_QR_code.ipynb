```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
image = cv2.imread("DATA/sample_qr.jpg", -1)
copy= image.copy()

qrDecoder = cv2.QRCodeDetector()
opencvData, bbox, rectifiedImage = qrDecoder.detectAndDecode(image)
# print("opencvData:",opencvData)

start_point= tuple(bbox[0][0])
end_point= tuple(bbox[0][2])

image= cv2.rectangle(image, start_point, end_point, color=(255,0,0), thickness=2) 
image= cv2.putText(image,opencvData,(10,40), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=0.3, color=(255,0, 0), thickness=1, lineType=cv2.LINE_AA)

plt.figure(figsize=[20,20]);
plt.subplot(121);plt.imshow(copy[:,:,::-1]); plt.title("Original Image");
plt.subplot(122);plt.imshow(image[...,0:3][...,::-1]);plt.title("Scanned");

plt.show()
```
